```python
#Import libraries 
import json
import glob
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

# Disease-Gene-SmallMolecule

### Distribution gene counts

```python
#Import result files
my_dir= "../results/Disease_SmallMolecule_Gene_/"

file_list = glob.glob(my_dir + "*")
```

```python
#Count number of genes in responses  

list_count = []

for file in file_list:
    print (file)
    with open (file) as f:
        data = json.load(f)
        
    count_genes= 0
    if "Error_Type:" in data:
        None


    if "message" in data:
        for rec in data["message"]["knowledge_graph"]["nodes"].values():
            for val in (rec["attributes"][0]["value"]):
                if "NCBIGene" in val:
                    count_genes+= 1  
        print (count_genes)
        list_count.append(count_genes)
```

```python
#Delete 0 (false results. Pairs that doesn't have genes)
my_count = [count for count in list_count if count != 0]
```

```python
#How many files have genes 
len(my_count)
```

```python
#Mean number genes in responses 
np.mean(my_count)
```

```python
#Maximun number of genes 
np.max(my_count)
```

```python
x = my_count

# the histogram of the data
sns.displot(x, bins=20, log= True)
plt.xlim((1,1000))
plt.ylim((1,1000))
plt.xlabel("Number of Genes")
plt.ylabel("Disease-Compound pairs")
plt.savefig('genes intermediate', dpi = 300)
```
