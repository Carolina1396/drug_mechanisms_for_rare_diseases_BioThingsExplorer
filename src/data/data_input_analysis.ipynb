### Count MONDO,Chembl

```python
#Import file
#Print oopd_dict saved file 
a_file = open("mondo_chembl_id.json",)
data = json.load(a_file)
```

```python
#Count number of MONDO
mondo = []

for lis in data:
    for item in lis.values():
        if "MONDO:" in str(item):
            mondo.append(item)
len(mondo)
```

```python
#Count number of CHEMBL
chem = []

for lis in data:
    for item in lis.items():
        if "CHEMBL" in item:
            for chembl in item[1]:
                if "CHEMBL" in str(chembl): 
                    chem.append(chembl)

#Unique chembl values 
values, counts = np.unique(chem, return_counts=True) 
```

```python
len(counts)
```
